```python
import pandas as pd
```

```python
df = pd.read_csv("samplesuperstorecsv.csv")
```

```python
df
```

```python
df = df.drop(columns=['Row ID','Product ID','Customer ID','Order ID'])
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(df['Sales'],bins=50,kde=True)
plt.title('Sales Distribution')
```

```python
sns.histplot(df['Profit'],bins=50,kde=True)
plt.title("Profit distribution")
```

```python
sns.boxplot(df['Quantity'])
plt.title('quantity spread')
plt.show()
```

```python
sns.scatterplot(x='Sales',y='Profit',data=df)
```

```python
df
```

```python
top_10 = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
top_10
```

```python
top_10_profit = df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(10)
```

```python
top_10_profit
```

```python
profit_cat = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
```

```python
df.groupby('Product Name')['Sales'].max().sort_values(ascending=False)
```

```python
# Which region has highest sales
```

```python
df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

# West has the highest sales
```

```python
# Which states are generating losses
df.groupby('State')['Profit'].sum().sort_values(ascending=True).head(12)
```

```python
# Which region has highest no of orders
df.groupby('Region')['Sales'].count().sort_values(ascending=False)
```

```python
# Top 10 customers by sales

df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False)
```

```python
# customers placing frequent small orders

customer_orders = df.groupby('Customer Name').agg({'Sales':['count','mean']})

```

```python
customer_orders
```

```python
# Average quantity sold

df.groupby('Product Name')['Quantity'].mean().sort_values(ascending=False)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce', infer_datetime_format=True)
```
